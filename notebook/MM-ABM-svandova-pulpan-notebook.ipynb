## Simulace na mřížce

```python
c1 = Culture(16, 1, 5, 5, 50000)
c1.plot_net()
c1.plot_gsnet()
```

```python
c1.simulate(save_progress=200)
c1.analyze()

print('200 iterací')
i = Image(filename='sim1_200.png'); display(i)
print('400 iterací')
i = Image(filename='sim1_400.png'); display(i)
print('600 iterací')
i = Image(filename='sim1_600.png'); display(i)
print(f'{c1.usediter} iterací')
i = Image(filename='sim1_final.png'); display(i)
```
